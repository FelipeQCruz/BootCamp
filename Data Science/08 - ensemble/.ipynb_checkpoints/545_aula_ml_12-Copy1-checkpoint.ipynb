# Outro olhar sobre kernel tricks SVM

Exemplos ilustrativos.

```python
import matplotlib
from matplotlib.colors import ListedColormap

def plot_fronteiras_de_decisao(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map up for up to 5 classes
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # Plot samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, color=cmap(idx),
                    marker='x', label=cl)

    # Plot test samples if they exist
    if not test_idx is None:
        X_test, y_test = X[test_idx, :], y[test_idx]
        for idx, cl in enumerate(np.unique(y_test)):
            plt.scatter(x=X_test[y_test == cl, 0], y=X_test[y_test == cl, 1],
                        alpha=1, color=cmap(idx),
                        linewidths=1, marker='o', s=55, label=f'test {cl}')
    plt.legend()
```

## Polynomial kernels

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Construir um dataset aleatório, a princípio

```python
# numero de pontos
n = 500

# assegurar reprodutibilidade
np.random.seed(13)

# dataset
X = np.random.randn(n, 2)
plt.scatter(X[:,0], X[:,1]);
```

### Criar o target apropriado

```python
noise = 0.2

epsilon = noise * np.random.randn(n, )
x1 = X[:, 0]
x2 = X[:, 1]

y_bool = np.logical_xor(
    x1 > epsilon, 
    x2 > epsilon
    )
y = np.where(y_bool, 1,0)

# verificar o dataset
plt.scatter(X[:, 0], X[:, 1], c=y)
```

### Modelos lineares

**Exercício:** Treine um modelo linear de sua escolha e use a função `plot_fronteiras_de_decisao` para verificar gráficamente a qualidade do resultado.

```python
from sklearn.svm import SVC

svm = SVC(C=1, kernel='linear')

svm.fit(X, y)

plot_fronteiras_de_decisao(X, y, classifier=svm, test_idx=None, resolution=0.02)
```

Qual a acurácia desse modelo?

```python
from sklearn.metrics import accuracy_score

y_pred = svm.predict(X)
accuracy_score(y, y_pred)
```

```python
from sklearn.linear_model import LogisticRegression

svm = LogisticRegression()
svm.fit(X, y)

plot_fronteiras_de_decisao(X, y, classifier=svm, test_idx=None, resolution=0.02)

y_pred = svm.predict(X)
accuracy_score(y, y_pred)
```

### Polynomial kernel

Mais detalhes [aqui](https://towardsdatascience.com/the-kernel-trick-c98cdbcaeb3f).

```python
from sklearn.svm import SVC

svm = SVC(C=10, degree=2, kernel='poly')

svm.fit(X, y)

plot_fronteiras_de_decisao(X, y, classifier=svm, test_idx=None, resolution=0.02)
```

```python
y_pred = svm.predict(X)
accuracy_score(y, y_pred)
```

### RBF kernel

```python
# outro dataset
from sklearn.datasets import make_moons
n = 200
X,y = make_moons(n_samples=n, noise=0.25, random_state=0)
plt.scatter(X[:,0], X[:,1], c=y)
```

# Métodos de Ensemble


## Instabilidade

```python
# condensando o exercicio da aula 9

import pandas as pd
import string

df = pd.read_csv('spam_ham.csv')



def clean_text(df_input, column_to_clean):
    
    df_input['text_clean'] = df_input[column_to_clean].str.lower()
    
    remover = f"[{string.punctuation}]"
    df_input['text_clean'] = df_input['text_clean'].str.replace(remover, '', regex=True)
    
    numeros = f"[1234567890]"
    df_input['text_clean'] = df_input['text_clean'].str.replace(numeros, '', regex=True)
    
    return df_input['text_clean']

df['text_clean'] = clean_text(df, 'text')

df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

def create_train_test_sets(df, semente):
    X = df['text_clean'].copy()
    y = df['type'].map({'ham':0, 'spam': 1})


    X_treino_bruto, X_teste_bruto, y_treino_, y_teste_ = train_test_split(X, y, 
                                                                          test_size=0.3, 
                                                                          random_state=semente)

    bag_of_words = CountVectorizer(max_features=100)

    X_treino_ = bag_of_words.fit(X_treino_bruto)
    X_treino_ = bag_of_words.transform(X_treino_bruto)
    X_teste_ = bag_of_words.transform(X_teste_bruto)
    
    return X_treino_, X_teste_, y_treino_, y_teste_


X_treino, X_teste, y_treino, y_teste = create_train_test_sets(df, 13)
```

```python
def evaluate_model(df, modelo, scoring_function, semente):
    
    X_treino, X_teste, y_treino, y_teste = create_train_test_sets(df, semente)
    
    modelo.fit(X_treino, y_treino)
    
    y_pred = modelo.predict_proba(X_teste)
    
    score = scoring_function(y_teste, y_pred[:,1])
    
    return score
    
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import roc_auc_score

mnb = MultinomialNB()

evaluate_model(df, mnb, roc_auc_score, 13)
```

```python
import seaborn as sns

sementes = [i for i in range(15)]

scores = [evaluate_model(df, mnb, roc_auc_score, i) for i in range(15)]


sns.lineplot(sementes, scores)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=13)

evaluate_model(df, tree, roc_auc_score, 13)
```

```python
sementes = [i for i in range(15)]

tree = DecisionTreeClassifier(max_depth=8, random_state=13) 

scores = [evaluate_model(df, tree, roc_auc_score, i) for i in sementes]


sns.lineplot(sementes, scores)

print(np.mean(scores),np.std(scores))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

evaluate_model(df, lr, roc_auc_score, 13)
```

```python
sementes = [i for i in range(15)]

lr = LogisticRegression()

scores = [evaluate_model(df, lr, roc_auc_score, i) for i in sementes]


sns.lineplot(sementes, scores)

print(np.mean(scores),np.std(scores))
```

## Bagging
Um meio de evitar a instabilidade é treinar diversos modelos **em paralelo** com **amostras** dos dados (técnia de **bootstrapping**) e combinar a decisão de todos eles no final. No caso de *regressão* fazemos a **média** dos resultados e para *classificação*, uma **votação**.

<img src="bagging_sketch.png">

```python
from sklearn.ensemble import BaggingClassifier

model = DecisionTreeClassifier(max_depth=10)

bg = BaggingClassifier(base_estimator=model, n_estimators=50, random_state=13)

bg.fit(X_treino, y_treino)

y_pred = bg.predict(X_teste)

roc_auc_score(y_teste, y_pred)
```

```python

sementes = [i for i in range(30)]

model = DecisionTreeClassifier(max_depth=10)

bg = BaggingClassifier(base_estimator=model, n_estimators=100)

scores = [evaluate_model(df, bg, roc_auc_score, i) for i in sementes]


sns.lineplot(sementes, scores)

print(np.mean(scores),np.std(scores))
```

```python
sementes = [i for i in range(30)]

model = LogisticRegression()

bg = BaggingClassifier(base_estimator=model, n_estimators=10)

scores = [evaluate_model(df, bg, roc_auc_score, i) for i in sementes]


sns.lineplot(sementes, scores)

print(np.mean(scores),np.std(scores))
```

## Boosting

Ao contrário do bagging, o método de boosting treina diversos modelos **em sequência**. 

<img src="boosting.png">




Essa técnica usa todos os dados e a cada iteração atribui diferentes pesos (importâncias) para os pontos que são classificados erroneamente (pode se tornar obcecado por outilers!).
