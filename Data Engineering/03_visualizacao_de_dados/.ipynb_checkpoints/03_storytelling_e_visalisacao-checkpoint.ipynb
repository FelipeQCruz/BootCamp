# Execcio

> 1. Crie uma query que d o total de itens vendidos, bem como a soma dos valores unitrios agrupados pelo nome do vendedor. 
> 2. Crie um filtro perodo. 
> 3. Reproduza os resultados em Power BI. 

## Resposta

`Select e.first_name, sum(od.quantity), sum(od.unit_price) 
from orders o
join order_details od on od.order_id = o.order_id
join employees e on e.employee_id = o.employee_id
where o.order_date between '1996-07-04' and '1998-05-06'
group by e.first_name
order by e.first_name`

## Conectando ao Power BI

```python
from powerbiclient import Report, models
```

```python
import requests
res = requests.get('https://playgroundbe-bck-1.azurewebsites.net/Reports/SampleReport')

access_token = res.json()['EmbedToken']['Token']
embed_url = res.json()['EmbedUrl']
```

```python
report=Report(access_token=access_token,embed_url=embed_url, token_type=models.TokenType.EMBED.value)
report
```
