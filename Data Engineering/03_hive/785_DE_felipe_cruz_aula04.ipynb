3. Edite o arquivo incluindo o conteúdo:

## Código

Todos os códigos abaixo foram executados pelo terminal.

### Preparando os arquivos no Hive
``
hdfs dfs -mkdir /user/hive/aula
vim aula.txt
hdfs dfs -put aula.txt /user/hive/aula
``

### Criação da tabela
``
beeline -u "jdbc:hive2://quickstart.cloudera:10000/default
``
```mysql
create table WordCount(word string);
load data inpath '/user/hive/aula/aula.txt' overwrite into table WordCount;
```
### Query para contagem das palavras
```mysql
select palavra, count(palavra)
from wordcount 
lateral view explode(split(word, ' ')) word as palavra
group by palavra;
```
